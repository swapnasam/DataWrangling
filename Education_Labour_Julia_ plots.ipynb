```julia
LU = LU |> @filter(_.Year > 2004) |> DataFrame # To select rows
```

```julia
names(Edu)
```

```julia
# To rename columns
rename!(Edu,:Region => :Country)
names!(LU, [:Country,:c2,:c3,:Labour_Force,:Unemployment_rate,:c6,:c7]) # Reference:https://syl1.gitbook.io/julia-language-a-concise-tutorial/useful-packages/dataframes
```

```julia
# To remove columns
select!(LU, Not(:2, :6, :7))
select!(Edu, Not(:1,:4, :5, :6, :7, :8, :9))  
```

```julia
names(LU)
```

```julia
Enr_lb = join(Edu,LU, kind =:inner , on = :Country, makeunique = true) # To join the DataFrames
```

```julia
describe(Enr_lb) # Structure of the dataframe
```

```julia
Enr_lb = Enr_lb |> @filter(_.Year == _.c3) |> DataFrame # To select rows
```

```julia
names(Enr_lb)
```

```julia
enr_lb = select!(Enr_lb, Not(:6))
```

```julia
enr_lb0 = enr_lb |> @filter(_.Country == ("New Zealand")) |> DataFrame
```

```julia
enr_lb1 = filter(r -> any(occursin.(["New Zealand", "Australia", "Canada"], r.Country)), enr_lb) # To select rows
# Reference: https://discourse.julialang.org/t/dataframes-obtaining-the-subset-of-rows-by-a-set-of-values/15923/5
```

```julia
# To rename all the columns
names!(enr_lb0, [:Country,:Year,:Primary_Enr,:Secondary_Enr,:Tertiary_Enr,:Labour_Force, :Unemployment_Rate]) 
```

```julia
names!(enr_lb1, [:Country,:Year,:Primary_Enr,:Secondary_Enr,:Tertiary_Enr,:Labour_Force, :Unemployment_Rate]) 
```
