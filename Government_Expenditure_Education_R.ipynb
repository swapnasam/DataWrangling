Dataset: Government_Expenditure_Education

Here, we are reading the file Government_Expenditure_Education.xls which is obtained from which contains the Government expenditurw on Education in different countries(http://data.uis.unesco.org/))

```r
library(tidyverse)
library(readxl)# to read excel
library(visdat) #for missing values
```

```r
# to read the excel Government_Expenditure_Education.xls by skipping the first three lines.
GEE <- read_excel('Government_Expenditure_Education.xls', col_names = TRUE, skip = 3)
GEE
```

```r
# We need from the year 2004 to 2018 only. Therefore removing other columns
GEE <- GEE[,-(3:48)]
GEE <- GEE[, -(17)]
```

```r
GEE %>% vis_miss()
GEE[is.na(GEE)] <- 'no data' # filling the missing values with NA
```

```r
GEE %>% vis_miss()
```

```r
names(GEE)[1] <- "Country"  #Renaming the country name to Country
GEE
```

```r
write_csv(GEE, path = "Government_Expenditure_Education_Wrangled.csv") # To save as CSV
```
