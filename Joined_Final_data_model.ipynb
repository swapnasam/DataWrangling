The final Data Model

Our target is to create a model which is a combination of New Zealand data from all 4 datasets (Expenditure, Enrolment, Labour and Unemployment). This is a relational data which shows just New Zealand information so we can compare the change over time of the country’s expenditure, enrolment, labour and unemployment

```r
library(tidyverse)
```

```r
#Read all the three wrangled files.

Edu_enroll <- read_csv("Education_enrollment_Wrangled.csv")
Govt_Expend <- read_csv("Government_Expenditure_Education_Wrangled.csv")
Labour_Unemployment <- read_csv("Labour__Unemployment_Rate_Wrangled.csv")
```

```r
Edu_enroll
Govt_Expend
Labour_Unemployment
```

```r
#To retrieve only the data for NZ from the enrollemnt table

NZ_Enrollemnt <- Edu_enroll %>%
  rename("Country" = "Countries") %>%
  filter(Country == "New Zealand") %>%
  select("Year", "Primary_enrollment")
  

 
NZ_Enrollemnt$Year <- as.integer(NZ_Enrollemnt$Year)
NZ_Enrollemnt<- NZ_Enrollemnt %>%
arrange(order(Year))
```

```r
NZ_Enrollemnt
```

```r
#To retrieve only the data for NZ from the govenrment expenditure table

NZ_Govt_Expend <- Govt_Expend %>%
  filter(Country == "New Zealand")
NZ_Govt_Expend <- NZ_Govt_Expend %>%
  gather(key=Year, value=PublicExpenditure)
NZ_Govt_Expend <- NZ_Govt_Expend[-(1:2),]
names(NZ_Govt_Expend)[2] <- "Government_Expenditure"

NZ_Govt_Expend$Year <- as.integer(NZ_Govt_Expend$Year)
```

```r
NZ_Govt_Expend
```

```r
#To retrieve only the data for NZ from the labour and unemployemnt table table
#Renaming the columns
Labour_Unemployment <- Labour_Unemployment %>%
  rename(Country = "Country Name", TotalUnemployment = "Unemployment, total (% of total labor force)", LabourForce ="Labor force, total") 
  
Labour_Unemployment$Year <- as.integer(Labour_Unemployment$Year)

```

```r
Labour_Unemployment
```

```r
NZ_Labour_Unemployment <- Labour_Unemployment %>%
  filter(Country == "New Zealand")%>%
  select(Year,LabourForce,TotalUnemployment)
```

```r
NZ_Labour_Unemployment
```

```r
##Joining of the three table using Year as the Key to join.

#NZ_Labour_Unemployment ,NZ_Govt_Expend and NZ_Enrollemnt

join_enroll_labour1 <- NZ_Govt_Expend %>%
  left_join(NZ_Labour_Unemployment,by ="Year")


Joined_df1 <- NZ_Enrollemnt %>%
  left_join(join_enroll_labour1,by ="Year")

```

```r
Joined_df1 # our  final data model obtained by joined the three tables.
```

```r
par(mfrow=c(1, 3))
boxplot(Joined_df1$Primary_enrollment, data = Joined_df1, 
        main = "Student enrolled primary",
        notch = F, 
        varwidth = TRUE, 
        col = c("green")
)
boxplot(Joined_df1$LabourForce , data = Joined_df1, 
        main = "Labour Force ",
        notch = F, 
        varwidth = TRUE, 
        col = c("green")
)
boxplot(Joined_df1$TotalUnemployment , data = Joined_df1, 
        main = "Total Unemployment ",
        notch = F, 
        varwidth = TRUE, 
        col = c("green")
)
```

The first plot shows that the median of students enrolled in primary level is 360, the minimal is under 350 and maximal is over 365. The second plot shows that the median of labour force is between 2,400,000 and 2,500,000, the minimal is under 2200000 and maximal is roughly 2700000. The third plot shows that the median of total unemployment is between 5.0 and 5.5, the minimal is under 4.0 and maximal is over 6.5. All boxplots does not have any outlier.

```r
ggplot(data = Joined_df1, aes(x = Year, y = TotalUnemployment)) +
  geom_point() + geom_line(linetype = "dashed")+
  labs(x = "Year",
       y = "Total unemployment",
       title = "Total unemployment in New Zealand")
```

#### From the time series plot we can see that the unemployment rate in New Zealand rises sharply from 2005 to 2010, and goes down until 2017.

```r
ggplot(data = Joined_df1, aes(x = Year, y = LabourForce)) +
geom_line(linetype = "dashed")+
  geom_point() +
  labs(x = "Year",
       y = "Labour force",
       title = "Labour force in New Zealand")
```

#### The labour force in New Zealand increases from 2005 to 2017 sharper and sharper.

```r
ggplot(data = Joined_df1, aes(x = Year, y = Primary_enrollment)) +
  geom_point() + geom_line(linetype = "dashed")+
  labs(x = "Year",
       y = "Student enrolled primary",
       title = "Student enrolled primary in New Zealand")
```

#### The student enrollment in primary in New Zealand goes down from 2005 to 2010, then increases rapidly.

```r
ggplot(data = Joined_df1, aes(x = Year, y = Government_Expenditure, group=1)) +
geom_line(linetype = "dashed")+
  geom_point() + 
  labs(x = "Year",
       y = "Government expenditure",
       title = "Government expenditure in New Zealand")
```

#### The Government expenditure in primary in New Zealand goes down sharply from 2005 to 2010, then increases rapidly
